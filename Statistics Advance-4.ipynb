Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

 The main difference between a t-test and a z-test lies in the information available about the population variance. A t-test is used when the population variance is unknown or estimated from the sample, whereas a z-test is used when the population variance is known.

For example, suppose you want to compare the mean heights of two groups of students (Group A and Group B). If you have the population variance of heights, you can use a z-test to compare the means. However, if you only have the sample variances of heights from each group, you would need to use a t-test.

Q2: Differentiate between one-tailed and two-tailed tests.

One-tailed and two-tailed tests refer to the directionality of the statistical hypothesis being tested.

In a one-tailed test, the hypothesis is specified to test for a difference or relationship in only one direction. For example, you may want to test if a new drug improves test scores and hypothesize that the drug will increase test scores. In this case, you would use a one-tailed test and look for evidence to support the hypothesis that the drug increases test scores.

In a two-tailed test, the hypothesis is not specific to a particular direction and tests for a difference or relationship in both directions. For example, you may want to test if there is a difference in test scores between two groups without specifying whether one group is expected to have higher scores. In this case, you would use a two-tailed test and look for evidence of a difference in either direction.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Type 1 error refers to rejecting a null hypothesis when it is actually true. It is the incorrect rejection of a true statement. An example scenario would be conducting a drug effectiveness study and concluding that the drug is effective when it actually has no effect.

Type 2 error refers to failing to reject a null hypothesis when it is actually false. It is the failure to reject a false statement. An example scenario would be conducting a drug effectiveness study and failing to conclude that the drug is effective when it actually has an effect.

Q4: Explain Bayes's theorem with an example.

Bayes's theorem is a fundamental concept in probability theory that describes how to update the probability of a hypothesis based on new evidence. It involves combining prior knowledge or beliefs (prior probability) with observed data (likelihood) to calculate the updated probability (posterior probability) of the hypothesis.

For example, suppose you want to know the probability of having a certain disease given a positive test result. Bayes's theorem allows you to combine the prevalence of the disease in the population (prior probability) with the accuracy of the test (likelihood) to calculate the updated probability of having the disease (posterior probability).

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It provides an estimate of the precision or uncertainty associated with a sample statistic.

To calculate a confidence interval, you need the sample statistic (e.g., mean or proportion), the sample size, the variability of the data, and the desired level of confidence. For example, if you want to estimate the mean height of a population with a 95% confidence interval, you would calculate the mean and standard deviation from your sample and use a formula to determine the range of values within which the true population mean is likely to fall with 95% confidence.

Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.

Let's assume we want to calculate the probability of a patient having a certain disease given the results of a diagnostic test.

Prior Information:

The prevalence of the disease in the population is 10% (prior probability).

The sensitivity of the diagnostic test is 90%, meaning it correctly identifies 90% of the patients who have the disease.

The specificity of the test is 95%, meaning it correctly identifies 95% of the patients who do not have the disease.

New Evidence:

A patient tests positive for the disease.

Using Bayes' Theorem, we can calculate the updated probability (posterior probability) of the patient having the disease given the positive test result.

Let:

P(D) = Probability of having the disease (prior probability) = 0.10

P(D') = Probability of not having the disease = 1 - P(D) = 0.90

P(T|D) = Probability of a positive test result given the disease (sensitivity) = 0.90

P(T|D') = Probability of a positive test result given no disease (1 - specificity) = 0.05

Now, we can calculate the posterior probability using Bayes' Theorem:

P(D|T) = (P(T|D) * P(D)) / [P(T|D) * P(D) + P(T|D') * P(D')]

P(D|T) = (0.90 * 0.10) / [(0.90 * 0.10) + (0.05 * 0.90)]

P(D|T) = 0.900 / (0.900 + 0.045)

P(D|T) ≈ 0.952 (or 95.2%)

Therefore, given a positive test result, the probability of the patient having the disease is approximately 95.2%.

Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5. Interpret the results.

To calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5, we need to consider the sample size and the desired level of confidence.

Using the formula for a confidence interval for a mean:

Confidence Interval = Mean ± (Critical Value * Standard Error)

Assuming a normal distribution and a sample size of 30 (for simplicity), we can find the critical value for a 95% confidence level from a standard normal distribution table (e.g., Z-table) or use a Z-score of 1.96 (corresponding to a 95% confidence level).

Standard Error = Standard Deviation / sqrt(Sample Size) = 5 / sqrt(30) ≈ 0.9129

Confidence Interval = 50 ± (1.96 * 0.9129)

Confidence Interval ≈ (48.18, 51.82)

Interpretation: With 95% confidence, the true population mean is estimated to fall within the range of 48.18 to 51.82.

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

The margin of error in a confidence interval represents the maximum amount by which a point estimate (e.g., mean) is expected to differ from the true population parameter.

The margin of error is influenced by the sample size. A larger sample size leads to a smaller margin of error because it reduces the variability and increases the precision of the estimate.

For example, if we conduct a survey to estimate the proportion of people in a city who support a particular policy, a larger sample size will provide a more accurate estimate of the true population proportion and, consequently, a smaller margin of error.

Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.

To calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5, we use the formula:

z = (x - μ) / σ

where x is the data point, μ is the population mean, and σ is the population standard deviation.

z = (75 - 70) / 5 = 1

Interpretation: The data point of 75 is 1 standard deviation above the population mean of 70.

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence level using a t-test, we need the sample mean, sample standard deviation, sample size, and the null hypothesis.

Null hypothesis (H0): The drug has no significant effect on weight loss.

Alternative hypothesis (Ha): The drug has a significant effect on weight loss.

Sample mean = 6 pounds

Sample standard deviation = 2.5 pounds

Sample size = 50

Significance level (α) = 0.05

We calculate the t-value using the formula:

t = (Sample Mean - Population Mean) / (Sample Standard Deviation / sqrt(Sample Size))

t = (6 - 0) / (2.5 / sqrt(50)) ≈ 10.66

We compare the calculated t-value with the critical t-value from a t-distribution table or software. If the calculated t-value exceeds the critical t-value, we reject the null hypothesis in favor of the alternative hypothesis.

Based on the degrees of freedom and significance level, we find that the critical t-value for a one-tailed test (assuming a positive effect) at a 95% confidence level is approximately 1.676.

Since the calculated t-value (10.66) is greater than the critical t-value (1.676), we reject the null hypothesis and conclude that the weight loss drug has a significant effect on weight loss at a 95% confidence level.

Confidence Interval = Sample Proportion ± (Critical Value * Standard Error)

Sample Proportion = 0.65

Sample Size = 500

Standard Error = sqrt((Sample Proportion * (1 - Sample Proportion)) / Sample Size)

Standard Error = sqrt((0.65 * (1 - 0.65)) / 500) ≈ 0.0202

Assuming a normal distribution, we can find the critical value for a 95% confidence level from a standard normal distribution table (e.g., Z-table) or use a Z-score of 1.96.

Confidence Interval = 0.65 ± (1.96 * 0.0202)

Confidence Interval ≈ (0.6101, 0.6899)

Interpretation: With 95% confidence, the true proportion of people who are satisfied with their job is estimated to fall within the range of 0.6101 to 0.6899.
