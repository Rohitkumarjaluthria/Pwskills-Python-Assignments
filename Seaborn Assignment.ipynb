Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
    
Seaborn is a powerful data visualization library in Python that provides various types of plots to analyze and explore data. Here are five commonly used plots in Seaborn along with their uses:

Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two continuous variables.
Seaborn Function: scatterplot()
Line Plot:

Use: Line plots are used to display the trend or change in a variable over continuous intervals, such as time.
Seaborn Function: lineplot()
Bar Plot:

Use: Bar plots are useful for comparing values across different categories or groups.
Seaborn Function: barplot()
Histogram:

Use: Histograms are used to visualize the distribution of a single variable.
Seaborn Function: histplot()
Heatmap:

Use: Heatmaps are useful for displaying the correlation or relationship between multiple variables in a tabular format.
Seaborn Function: heatmap()

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Plot line plot using Seaborn
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Plot box plots using Seaborn
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.show()

sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

diamonds = sns.load_dataset("diamonds")

sns.histplot(data=diamonds, x='price', hue='cut')

plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Plot pair plot using Seaborn
sns.pairplot(data=iris, hue='species')

# Display the pair plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix
flights_matrix = flights.pivot("month", "year", "passengers")

# Plot a heatmap using Seaborn
sns.heatmap(flights_matrix, cmap="YlGnBu" )

# Display the heatmap
plt.show()
```
