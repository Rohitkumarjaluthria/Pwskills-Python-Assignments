Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.

The Probability Mass Function (PMF) and Probability Density Function (PDF) are statistical functions used to describe the probability distribution of a discrete random variable (PMF) or a continuous random variable (PDF).

The PMF calculates the probability of a discrete random variable taking on a specific value. It gives the probability of each possible outcome of the random variable. For example, consider a fair six-sided die. The PMF would assign a probability of 1/6 to each outcome (1, 2, 3, 4, 5, 6).

The PDF, on the other hand, gives the probability density at each possible value of a continuous random variable. It represents the relative likelihood of different outcomes occurring. For example, in the case of a normal distribution, the PDF describes the shape of the bell curve.

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

The Cumulative Density Function (CDF) is a function that gives the probability that a random variable takes on a value less than or equal to a given value. It gives the cumulative probability up to a certain point.

For example, in the case of a normal distribution, the CDF gives the probability that a random variable falls within a certain range. It is calculated by integrating the PDF of the normal distribution. The CDF is a monotonically increasing function, starting from 0 at negative infinity and reaching 1 at positive infinity.

The CDF is used to calculate various statistical quantities, such as percentiles and confidence intervals. It provides a way to analyze the likelihood of a random variable falling within a specific range or threshold.

The normal distribution is commonly used as a model in situations where data tends to cluster around a central value with symmetrically distributed values on both sides. Some examples of situations where the normal distribution might be used as a model include:

IQ scores

Errors in measurements

Natural phenomena such as the distribution of rainfall or wind speeds

The parameters of the normal distribution are the mean (μ) and standard deviation (σ). The mean determines the center of the distribution, while the standard deviation determines the spread or dispersion of the data. A smaller standard deviation corresponds to a narrower and taller bell curve, while a larger standard deviation results in a wider and flatter bell curve.

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

The Bernoulli distribution is a discrete probability distribution that models a single trial with two possible outcomes: success (usually represented by 1) and failure (usually represented by 0). It is named after the Swiss mathematician Jacob Bernoulli. The distribution is characterized by a single parameter, p, which represents the probability of success.

An example of the Bernoulli distribution is flipping a fair coin. The outcome of each flip can be either heads (success) or tails (failure). The probability of heads (p) is 0.5, and the probability of tails (1-p) is also 0.5.

Now, let's discuss the difference between the Bernoulli distribution and the Binomial distribution:

Bernoulli Distribution:

Models a single trial with two possible outcomes (success or failure).

Has a single parameter, p, representing the probability of success.

Probability mass function (PMF) is given by P(X=k) = p^k * (1-p)^(1-k), where k is the outcome (0 or 1).

Binomial Distribution:

Models the number of successes in a fixed number of independent trials.

Has two parameters: n (the number of trials) and p (the probability of success in each trial).

Probability mass function (PMF) is given by P(X=k) = C(n, k) * p^k * (1-p)^(n-k), where k is the number of successes and C(n, k) is the binomial coefficient.

Q7: Explain uniform Distribution with an example.

The uniform distribution is a probability distribution where all possible outcomes are equally likely. In other words, it represents a situation where each possible value has the same probability of occurring. The probability density function (PDF) of a uniform distribution is constant within a specified interval and zero outside that interval.

For example, let's consider a fair six-sided die. Each side has an equal chance of landing face up, and the probability of getting any specific number (1, 2, 3, 4, 5, or 6) is 1/6. This follows a uniform distribution because each outcome is equally likely.

Q8: What is the z score? State the importance of the z score.

The z-score, also known as the standard score, measures the distance of a data point from the mean of a distribution in terms of standard deviations. It is calculated by subtracting the mean from the data point and dividing the result by the standard deviation. The formula for calculating the z-score is: z = (x - μ) / σ

The importance of the z-score lies in its ability to standardize data, making it easier to compare values from different distributions. It allows us to determine how far away a data point is from the mean relative to the spread of the data. Z-scores are commonly used in statistical analysis, hypothesis testing, and determining confidence intervals.
