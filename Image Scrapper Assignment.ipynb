```python
#Q1. Write a python program to extract the video URL of the first five videos.


import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.youtube.com/@PW-Foundation/videos"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

video_urls = []
for i, video in enumerate(soup.find_all("video")):
    if i < 5:
        video_url = video.get("src")
        video_urls.append(video_url)

with open("video_urls.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Video URL"])
    writer.writerows([[url] for url in video_urls])
```

```python
#Q3. Extract the title of the first five videos:

import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.youtube.com/@PW-Foundation/videos"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

titles = []
for i, video in enumerate(soup.find_all("video")):
    if i < 5:
        title = video.get("title")
        titles.append(title)

with open("video_titles.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Title"])
    writer.writerows([[title] for title in titles])
```

```python
#Q4. Extract the number of views of the first five videos:

import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.youtube.com/@PW-Foundation/videos"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

views = []
for i, video in enumerate(soup.find_all("video")):
    if i < 5:
        views_count = video.get("views")
        views.append(views_count)

with open("video_views.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Views"])
    writer.writerows([[count] for count in views])
```

```python
#Q5. Extract the time of posting of video for the first five videos:

import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.youtube.com/@PW-Foundation/videos"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

post_times = []
for i, video in enumerate(soup.find_all("video")):
    if i < 5:
        post_time = video.get("post_time")
        post_times.append(post_time)

with open("video_post_times.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Post Time"])
    writer.writerows([[time] for time in post_times])
```
