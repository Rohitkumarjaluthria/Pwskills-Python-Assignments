
Q1. What is an API? Give an example where an API is used in real life.

An API (Application Programming Interface) is a set of rules and protocols that allows different software applications to communicate and interact with each other. It defines how software components should interact and the methods they can use to exchange data.

An example of API usage in real life is when you use a social media login feature on a website or application. The website or application doesn't store your login credentials itself; instead, it relies on an API provided by the social media platform (such as Facebook, Google, or Twitter) to authenticate and authorize your login. The website or application uses the API to securely communicate with the social media platform and retrieve the necessary user data for authentication and authorization purposes.

Q2. Give advantages and disadvantages of using API.

Advantages of using API:

Modularity: APIs promote modularity by allowing applications to be built in separate components that can be developed, updated, and maintained independently.
Reusability: APIs enable code reuse, as developers can leverage existing APIs to add functionalities to their applications without reinventing the wheel.
Scalability: APIs support scalability by allowing applications to interface with external services and resources, offloading specific tasks and distributing the workload.
Collaboration: APIs facilitate collaboration between developers and teams by providing clear boundaries and specifications for interaction between different software components.
Disadvantages of using API:

Dependency: APIs create a dependency on the provider, as any changes or disruptions in the API can impact the functionality and stability of dependent applications.
Complexity: Working with APIs can be complex, especially when dealing with large and comprehensive APIs with extensive documentation and various endpoints.
Security Risks: APIs can introduce security vulnerabilities if not implemented and secured properly. Unauthorized access, data breaches, and malicious attacks can occur if API security measures are inadequate.
Versioning and Compatibility: As APIs evolve and change over time, managing version compatibility between different API versions and ensuring smooth transitions can be challenging.

Q3. What is a Web API? Differentiate between API and Web API.

A Web API, also known as a Web Service API, is an API specifically designed to be used over the web using the HTTP protocol. It enables communication and interaction between different software systems and applications over a network.

The main difference between API and Web API is the medium of communication. APIs, in general, refer to a broader concept of defining interfaces between software components, which can be used for communication within a single application or between different applications or systems. On the other hand, Web APIs specifically focus on enabling communication between applications over the web using standard protocols like HTTP.

Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

REST (Representational State Transfer) and SOAP (Simple Object Access Protocol) are two different architectural styles for designing web services:

REST Architecture:

REST is an architectural style that uses standard HTTP methods (GET, POST, PUT, DELETE) to communicate with resources over the web.
It relies on a stateless client-server communication model, where each request from the client contains all the necessary information to be understood by the server.
REST APIs use lightweight data formats such as JSON or XML for data interchange.
REST APIs are designed to be simple, scalable, and easy to consume, making them widely adopted for web services.

SOAP Architecture:

SOAP is a protocol and messaging format used for exchanging structured information in web services.
It relies on XML for message formatting and uses additional XML-based protocols for messaging, security, and routing.
SOAP APIs have more rigid and complex specifications, including a WSDL (Web Services Description Language) for service definition and a SOAP envelope structure for message packaging.
SOAP provides additional features such as built-in error handling, security, and reliable messaging.

Shortcomings of SOAP:

Complexity: SOAP APIs are generally more complex to develop and maintain due to the extensive XML-based specifications and additional protocols involved.

Performance Overhead: SOAP can have higher overhead due to the verbosity of XML and the additional processing required for parsing and interpreting SOAP messages.

Limited Language Support: SOAP has limited language support compared to REST, as it is primarily based on XML and not as widely embraced by modern programming languages.

Firewall and Proxy Limitations: SOAP can face challenges with firewalls and proxies due to its use of non-standard HTTP headers and XML payload, making it less compatible with certain network configurations.
