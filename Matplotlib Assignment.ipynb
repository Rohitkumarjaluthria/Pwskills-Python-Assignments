Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating various types of plots and charts. It is widely used in the scientific and data analysis communities for visualizing data in a clear and meaningful way.

Matplotlib is used for tasks such as:

Exploratory Data Analysis (EDA): Matplotlib helps in visualizing data distributions, trends, and patterns, making it easier to understand the underlying structure and relationships within the data.

Presentation and Reporting: Matplotlib allows you to create high-quality, publication-ready visualizations that can be embedded in reports, papers, presentations, or shared online.

Customization and Styling: Matplotlib provides extensive customization options to fine-tune the appearance of plots, including control over colors, line styles, markers, fonts, labels, and more.

Plotting for Machine Learning: Matplotlib is often used for visualizing the results of machine learning algorithms, such as displaying loss curves, decision boundaries, or feature importance.

Interactivity: Matplotlib supports interactive features such as zooming, panning, and mouse hovering, allowing users to explore and interact with plots in real-time.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line Plot: A line plot is used to display data points connected by straight lines. It is commonly used to visualize trends and changes over time.

Scatter Plot: A scatter plot shows individual data points as markers on a coordinate grid. It is useful for visualizing the relationship between two variables or detecting patterns or outliers.

Bar Plot: A bar plot represents data with rectangular bars of varying heights. It is commonly used to compare categorical data or display counts, frequencies, or distributions.

Histogram: A histogram is used to visualize the distribution of a continuous variable. It displays the frequency of values within specific bins or ranges.

Pie Chart: A pie chart represents data as slices of a circular pie. It is useful for displaying proportions or percentages of a whole and is often used to compare categorical data.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Bar Plot')
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)


box2 = np.random.normal(90, 20, 200)

 A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the distribution of a dataset. It displays the minimum, first quartile, median, third quartile, and maximum values of a dataset, along with any potential outliers.

Box plots are used for the following purposes:

Distribution Summary: A box plot provides a visual summary of the central tendency (median) and spread (interquartile range) of a dataset. It allows for quick comparisons between multiple datasets or categories.

Outlier Detection: Box plots visually represent potential outliers in a dataset. Outliers are represented as individual points beyond the whiskers, providing insights into extreme values or anomalies.

Skewness and Symmetry: By examining the lengths and positions of the whiskers, box plots provide information about the skewness and symmetry of the data distribution. For symmetric distributions, the whiskers are of similar length, while for skewed distributions, the whiskers may differ in length.

Comparisons: Box plots are useful for comparing distributions across different groups or categories. They allow for the identification of differences in medians, quartiles, or spread between groups.


```python

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
